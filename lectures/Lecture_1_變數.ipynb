# 數字
程式語言都有計算的功能，python 當然也有

下面的範例是計算 100 元紙鈔購買 65 元飲料的找零為多少

````python
100 - 65 # 找零金額 (# 後面的文字是註解，不被執行)
````

code 下方顯示的數字就是計算結果，基本上和工程計算機是相似的

**但是程式語言可以定義變數 (variable)**

定義變數有幾個好處
* 知道那個值的意義
* 如果要修改數值的話只要修改定義的地方就好

---
#### Syntax (句法)
```python
price = 65
```
---


````python
payment = 100 # 付多少錢
price = 65    # 商品金額
change_give_back = payment - price # 找錢

change_give_back # 最後這行是讓 python 顯示 `change_give_back` 的值
````

變數名稱可以包含:
* 文字 (中文、希臘文也可以)
* 數字 (但不能在開頭)
* _ (在 `change_give_back` 的例子是代替空白，不然黏在一起很難看懂)

> 另外 `#` 在 python 裡面是註解的意思，在 `#` 後面的的文字都不會被 python 執行。

下面是一些不合法的變數名稱:

````python
1_var = 0
````

````python
change give back = 0
````

`SyntaxError` 指的是句法錯誤

**記得數字不能在文字前面，也記得用 `_` 代替空格哦**

#### 補充: code style
在 python 中慣例上會使用 `_` 代替空格只是一個慣例，來自於這份 python 官方的文件 [PEP 8](https://peps.python.org/pep-0008/)。

其他語言 (如: C, C++, javascript) 可能更偏好: `changeGiveBack`

或是你想寫 `changegiveback` 也當然可以，只是很難看懂啦。


## 運算符 (operator)
| 符號 | 意義 |
| --- | --- |
| + | 加法 |
| - | 減法 |
| * | 乘法 |
| / | 除法 |
| // | 商數 (只有整數) |
| % | 餘數 |
| ** | 次方 |

````python
eleven = 11 # 名字是為了讓大家知道他就是 11 或 2
two = 2

eleven + two
eleven - two
eleven * two
eleven / two
eleven // two
eleven % two
eleven ** two
````

**怎麼只有一行?**

只有最後一行的結果會顯示哦，所以如果我們不想開很多個 code cell，想要一次顯示的話要用 `print()` 這個函數。

只要在 `()` 裡面放你要的值就可以了。

````python
print(eleven + two) # 會先執行括號內的運算式後再丟給 print 去顯示數值
print(eleven - two)
print(eleven * two)
print(eleven / two)
print(eleven // two)
print(eleven % two)
print(eleven ** two)
````

# 一元二次方程式的解
最後再加一個複雜一點的例子
$$ a x ^ 2 + b x + c = 0$$
$$x = \frac{-b \pm \sqrt{b^2 - 4 ac}}{2a}$$

````python
a = 1
b = 2
c = -3

sqrt_part = (b ** 2 - 4 * a * c) ** 0.5 # 順序: 括號 -> 次方 -> 乘除 -> 加減

x_1 = (-b + sqrt_part) / (2 * a) # 當然還是先括號，再乘除後加減
x_2 = (-b - sqrt_part) / (2 * a)
print(x_1)
print(x_2)
````

## 運算符
字串也有運算符可以使用，但是只有 `+` 和 `*`
* `+`: 將字串與字串接起來
* `*`: 重複該字串許多次

# list 串列 (預告)

當我們的數字很多的時候就不適合一個一個宣告變數了，有一個容器把他們集中會更好管理，`list` 可以完成這個任務

---
#### Syntax

```pyhton
[1, 2, 3, "字串與數字裝在一起也可以"]

[
    1,
    2,
    "也可以這樣寫，雖然很占版面",
    "但是更好閱讀，改起來也方便"
]
```
---

**記得一樣要用逗號 `,` 將變數隔開**

````python
sample_list_1 = [1, 2.0, "我是字串"]
print(sample_list_1) # 一樣可以用 print 印出來

sample_list_2 = [
    "如果很長的話直著寫會比較清楚",
    "很多編輯器有刪除整行或移動整行的功能",
    "所以運用這些功能的話直著寫會更好改",
]
print(sample_list_2)
````

之後會進一步介紹這些容器的使用方式。但現在我們已經可以使用 python 內建的函式做些基本的統計了
* 計算總和 `sum()`
* 最大值 `max()`
* 最小值 `min()`
* 幾個變數 `len()` (可以用來求平均)

````python
food_expenses = [0, 75, 110] # 三餐的費用
print("6/24 餐費")
print("最高:", max(food_expenses)) # 直接把 list 塞進函式即可
print("最低:", min(food_expenses))
print("總計:", sum(food_expenses))
print("平均:", sum(food_expenses) / len(food_expenses))
````
